```python
# the following code can be used to downsample/upsample audio files
import os
import subprocess

input_path = r"/home/shh/ANewDisk/RyanSpeech/train_copy/"
output_path = r"/home/shh/ANewDisk/RyanSpeech/train_copy_downsampled/"
for file in os.listdir(input_path):
    file1 = input_path+'\\'+file
    file2 = output_path+'\\'+file
    cmd = "ffmpeg -i " + file1 + " -ar 16000 " + file2  #ffmpeg -i  -ar   
    subprocess.call(cmd, shell=True)
```
